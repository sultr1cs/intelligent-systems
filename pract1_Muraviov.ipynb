### 6. Тестові сценарії (4 сценарії)



```python
test_scenarios = [
    {"name": "Сценарій 1: Типовий багатий геймер (декілька спрацьовувань)",
     "facts": {"primary_goal": "gaming", "budget": "high", "mobility": "no", "priority": "performance"}},

    {"name": "Сценарій 2: Дизайнер-фан Apple (крайній випадок з ланцюжком)",
     "facts": {"primary_goal": "design_video", "budget": "high", "mobility": "yes", "os_preference": "macos"}},

    {"name": "Сценарій 3: Офісний працівник у відрядженнях",
     "facts": {"primary_goal": "office", "budget": "medium", "mobility": "yes", "priority": "portability"}},

    {"name": "Сценарій 4: Випадок без чіткого висновку щодо класу",
     "facts": {"primary_goal": "basic_home", "budget": "medium", "mobility": "no", "priority": "performance"}}
]

print("Тестові сценарії готові.")
```

### 7. Механізм логічного висновку та запуск сценаріїв



```python
def forward_chain(facts, rules):
    expanded = dict(facts)
    log = []
    new_facts_added = True

    while new_facts_added:
        new_facts_added = False
        for rule in rules:
            match = True
            for key, value in rule["conditions"].items():
                if expanded.get(key) != value:
                    match = False
                    break
            
            if match:
                k, v = rule["conclusion"]
                if expanded.get(k) != v:
                    expanded[k] = v
                    log.append(f"Спрацювало {rule['name']}: додано {k} = {v}")
                    new_facts_added = True

    return expanded, log

# Запуск перевірки на сценаріях
for scenario in test_scenarios:
    print(f"\n{'='*40}\n--- {scenario['name']} ---")
    print("Вхідні дані:", scenario["facts"])
    result, log = forward_chain(scenario["facts"], rules)
    print("\nЖурнал виконання:")
    for entry in log:
        print("  -", entry)
    print("\nФінальний результат:")

    for target in ["form_factor", "device_class", "recommended_os"]:
        if target in result:
            print(f"  > {target.capitalize()}: {result[target]}")
    if "device_class" not in result:
        print("  > УВАГА: Клас пристрою не знайдено (недостатньо правил або даних).")
```
